```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Open the image and convert it to grayscale
image_path = '/workspaces/Heart-CNN/data/angina_disease/angina_disease (1).jpe'
img = Image.open(image_path).convert('L')

# Convert the grayscale image to a NumPy array (matrix)
grayscale_matrix = np.array(img)

# Define a 3x3 kernel (for example, an edge-detection kernel)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution (dot product) over the image
filtered_image = convolve(grayscale_matrix, kernel)

# Convert the filtered matrix to a DataFrame and save it as a CSV
filtered_df = pd.DataFrame(filtered_image)
csv_path = '/workspaces/Heart-CNN/data/filtered_image.csv'
filtered_df.to_csv(csv_path, index=False)

# Display the original and filtered image
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(grayscale_matrix, cmap='gray')
ax[0].set_title('Original Grayscale Image')
ax[0].axis('off')

ax[1].imshow(filtered_image, cmap='gray')
ax[1].set_title('Filtered Image')
ax[1].axis('off')

plt.show()

print(f"Filtered image saved to: {csv_path}")
```

```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Open the image and convert it to grayscale
image_path = '/workspaces/Heart-CNN/data/angina_disease/angina_disease (1).jpe'
img = Image.open(image_path).convert('L')

# Convert the grayscale image to a NumPy array (matrix)
grayscale_matrix = np.array(img)

# Define a 3x3 kernel (for example, an edge-detection kernel)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution using convolve2d
full_convolution = convolve2d(grayscale_matrix, kernel, mode='same', boundary='wrap')

# Apply stride of 5 pixels
stride = 5
strided_filtered_image = full_convolution[::stride, ::stride]

# Convert the strided filtered matrix to a DataFrame and save it as a CSV
filtered_df = pd.DataFrame(strided_filtered_image)
csv_path = '/workspaces/Heart-CNN/data/filtered_image.csv'
filtered_df.to_csv(csv_path, index=False)

# Display the original and filtered image
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(grayscale_matrix, cmap='gray')
ax[0].set_title('Original Grayscale Image')
ax[0].axis('off')

ax[1].imshow(strided_filtered_image, cmap='gray')
ax[1].set_title('Filtered Image with Stride')
ax[1].axis('off')

plt.show()

print(f"Filtered image saved to: {csv_path}")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths to your dataset directory and the split directories
dataset_dir = '/workspaces/Heart-CNN/data'
output_dirs = {
    'train': os.path.join(dataset_dir, 'training'),
    'val': os.path.join(dataset_dir, 'validation'),
    'test': os.path.join(dataset_dir, 'testing')
}

# Create directories if they do not exist
for path in output_dirs.values():
    if not os.path.exists(path):
        os.makedirs(path)

# List of disease categories (folders)
categories = ['angina_disease', 'cardio_disease_', 'coronary_artery_disease', 'hypotension_disease_']

# Function to split data
def split_data(category):
    category_dir = os.path.join(dataset_dir, category)
    if not os.path.exists(category_dir):
        print(f"Category directory {category_dir} does not exist.")
        return

    # List all image files
    images = [os.path.join(category_dir, img) for img in os.listdir(category_dir) if img.lower().endswith(('jpeg', 'jpg', 'png'))]
    
    if not images:
        print(f"No images found in {category_dir}.")
        return

    # Split data
    train_val, test = train_test_split(images, test_size=0.2, random_state=42)
    train, val = train_test_split(train_val, test_size=0.2, random_state=42)
    
    # Ensure destination directories exist
    for folder in ['train', 'val', 'test']:
        dest_dir = os.path.join(output_dirs[folder], category)
        if not os.path.exists(dest_dir):
            os.makedirs(dest_dir)

    # Move files to the appropriate directories
    print(f"Moving files for category: {category}")
    for file in train:
        shutil.copy(file, os.path.join(output_dirs['train'], category))
    for file in val:
        shutil.copy(file, os.path.join(output_dirs['val'], category))
    for file in test:
        shutil.copy(file, os.path.join(output_dirs['test'], category))

# Split data for each category
for category in categories:
    split_data(category)

print("Data splitting and moving complete.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image parameters
img_height, img_width = 399, 379
batch_size = 32

# Data augmentation and preprocessing
datagen = ImageDataGenerator(rescale=1./255)

# Paths to your dataset directories
train_dir = '/workspaces/Heart-CNN/data/training'
val_dir = '/workspaces/Heart-CNN/data/validation'
test_dir = '/workspaces/Heart-CNN/data/testing'

# Training data generator
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=True
)

# Validation data generator
validation_generator = datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False
)

# Testing data generator
test_generator = datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False
)
```

```python
import matplotlib.pyplot as plt

# Get a batch of images and labels from the train_generator
batch = next(train_generator)
images, labels = batch

# Retrieve the first image and its corresponding label
image = images[0]
label = labels[0]

# Get class names from generator
class_names = list(train_generator.class_indices.keys())

# Find the class index of the label
label_index = tf.argmax(label).numpy()
class_name = class_names[label_index]

# Plot the image
plt.figure(figsize=(5, 5))
plt.imshow(image)
plt.title(f'Label: {class_name}')
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.applications import VGG16
from tensorflow.keras.optimizers import Adam

# Define the number of classes
num_classes = 4

# Load the VGG16 model pre-trained on ImageNet without the top (classification) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(399, 379, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Create the custom model on top of the base VGG16 model
x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dense(num_classes, activation='softmax')(x)  # Output layer for classification

# Define the model
model = Model(inputs=base_model.input, outputs=x)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()

# Define the number of epochs and steps per epoch
epochs = 10
steps_per_epoch = train_generator.samples // batch_size
validation_steps = validation_generator.samples // batch_size

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_steps
)

# Save the trained model
model.save('vgg16_heart_disease_model.h5')

# Evaluate the model on the test set
test_steps = test_generator.samples // batch_size

test_loss, test_accuracy = model.evaluate(
    test_generator,
    steps=test_steps
)

print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```
